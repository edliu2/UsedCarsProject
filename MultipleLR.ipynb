```python
#Multiple Linear Regression

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

#Loading the dataset
dataset = pd.read_csv('drive/My Drive/processed.csv')

dataset.head()
```

```python

X = dataset.iloc[:,2:].values 
y = dataset.iloc[:,1].values

#Splitting the data into Training Set and Test Set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3)

#Normalizing the features
from sklearn.preprocessing import StandardScaler

#Create and Fit a Regressor Object
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline

pipe_lr = Pipeline([('scale', StandardScaler()),
    ('lr', LinearRegression())
    ])

grid_params = {
    'lr__fit_intercept': [True, False],
    'lr__normalize': [True, False]
}

#Run grid search cross validated on pipeline
from sklearn.model_selection import GridSearchCV
resultObj = GridSearchCV(pipe_lr, param_grid=grid_params, n_jobs=-1, scoring='neg_mean_squared_error', verbose=10, cv=10)

resultObj.fit(X_train, y_train)
y_pred = resultObj.predict(X_test)

print(resultObj.cv_results_)
print(resultObj.best_params_)
```

```python
from sklearn.metrics import mean_squared_error
print(mean_squared_error(y_test, y_pred))
```
